```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.linspace(0,10,200)
```

```python
x.shape
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('this is sin curve')
plt.plot(x,y)
plt.show()
```

```python
y
```

```python
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('this is sin curve')
plt.plot(x,y,'--b')
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('this is sin curve')
plt.plot(x,y,'--r')
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('this is sin curve')
plt.plot(x,y,'--g')
plt.grid()
plt.show()
```

```python
y1= np.sin(x)
y2 = np.cos(x)
```

```python

plt.figure(figsize=(8,4))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('this is sin and cos curve')
plt.plot(x,y1,x,y2)
plt.grid()
plt.show()
```

## Scatter plot

```python
x= np.random.rand(30)
y=np.random.rand(30)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
color= np.random.rand(30)
plt.scatter(x,y,c=color)
plt.show()
```

```python
size= np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c=color,s=size)
plt.show()
```

```python
size
```

```python
size= 100*np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c=color,s=size)
plt.show()
```

```python
size
```

```python
size= 100*np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c=color,s=size,alpha=.7)
plt.show()
```

```python
size= 100*np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c=color,s=size,alpha=.7)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Data Graph')
plt.show()
```

```python
size= 100*np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c='#244B5F',s=size,alpha=.7)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Data Graph')
plt.show()
```

```python
size= 100*np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c=color,s=size,alpha=.7,marker='v')
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Data Graph')
plt.show()
```

```python
size= 100*np.random.rand(30)
color= np.random.rand(30)
plt.scatter(x,y,c=color,s=size,alpha=.7,marker='a')
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Data Graph')
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df.head()
```

```python
df[['Pclass']].value_counts()
```

```python
g= df.groupby('Pclass')['Fare'].sum().reset_index()
print(g)
```

```python
plt.bar(g['Pclass'], g['Fare'])
```

```python
g1= df.groupby('Sex')['Fare'].sum().reset_index()
print(g1)
```

```python
plt.bar(g1['Sex'], g1['Fare'])
```

```python
plt.barh(g1['Sex'], g1['Fare'])
```
